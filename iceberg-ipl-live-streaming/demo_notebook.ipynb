# Functions to Fetch IPL Information

```python
import requests
from config import *
import pandas as pd
from datetime import datetime
```

```python
def series_info(series,api_key):
    url = f"https://api.cricapi.com/v1/series_info?apikey={api_key}&id={series}"
    response = requests.get(url)
    data = []
    if response.status_code == 200:
        data = response.json()
        matches = data["data"]["matchList"]
        df = pd.DataFrame(matches)
        csv_file = f"./data/series_data_{datetime.now().strftime('%Y-%m-%d')}.csv"
        df.to_csv(csv_file, index=False)
    else:
        print(f"Error: {response.status_code}, {response.text}")
    return df

series_df = series_info(series_id,api_key_1)
series_df

def series_squad_info(series,api_key):
    url = f"https://api.cricapi.com/v1/series_squad?apikey={api_key}&id={series}"
    response = requests.get(url)
    data = []
    if response.status_code == 200:
        data = response.json()
        matches = data["data"]
        df = pd.DataFrame(matches)
        csv_file = f"./data/series_squad_data_{datetime.now().strftime('%Y-%m-%d')}.csv"
        df.to_csv(csv_file, index=False)
    else:
        print(f"Error: {response.status_code}, {response.text}")
    return df

series_squad_df = series_squad_info(series_id,api_key_1)
series_squad_df

# Function to get the match details {historical | current | all}
def match_info(api_key,dataframe, load):
    matches_data = []
    if load == 'historical':
        match_ids = dataframe[dataframe["date"] <= datetime.now().strftime('%Y-%m-%d')]["id"].to_list()
    elif load == 'current':
        match_ids = dataframe[dataframe["date"] == datetime.now().strftime('%Y-%m-%d')]["id"].to_list()
    elif load == 'current':
        match_ids = dataframe[dataframe["date"] > datetime.now().strftime('%Y-%m-%d')]["id"].to_list()
    else:
        match_ids = dataframe[dataframe["date"] == datetime.now().strftime('%Y-%m-%d')]["id"].to_list()
    for match_id in match_ids:
        url = f"https://api.cricapi.com/v1/match_info?apikey={api_key}&offset=0&id={match_id}"
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            matches = data["data"]
            matches_data.append(matches)
        else:
            print(f"Error: {response.status_code}, {response.text}")    
    df = pd.DataFrame(matches_data)
    csv_file = f"./data/matches_data_{datetime.now().strftime('%Y-%m-%d')}.csv"
    df.to_csv(csv_file, index=False)
    return df

matches_df  = match_info(api_key_2,series_df,'historical')
matches_df

def match_bbb_info(api_key,dataframe):
    matches_data = []
    match_ids = dataframe[dataframe["date"] < datetime.now().strftime('%Y-%m-%d')]["id"].to_list()
    for match_id in match_ids:
        url = f"https://api.cricapi.com/v1/match_bbb?apikey={api_key}&id={match_id}"
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            matches = data["data"]
            matches_data.append(matches)
        else:
            print(f"Error: {response.status_code}, {response.text}")    
    df = pd.DataFrame(matches_data)
    csv_file = f"./data/matches_data_bbb_{datetime.now().strftime('%Y-%m-%d')}.csv"
    df.to_csv(csv_file, index=False)
    return df
matches_bbb_df = match_bbb_info(api_key_2,matches_df)

def match_squad_info(api_key,dataframe):
    matches_data = []
    match_ids = dataframe[dataframe["date"] < datetime.now().strftime('%Y-%m-%d')]["id"].to_list()
    for match_id in match_ids:
        url = f"https://api.cricapi.com/v1/match_squad?apikey={api_key}&id={match_id}"
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            matches = data["data"]
            matches[0]["match_id"] = f"{match_id}"
            matches[1]["match_id"] = f"{match_id}"
            matches_data.append(matches)
        else:
            print(f"Error: {response.status_code}, {response.text}")
    df = pd.DataFrame(matches_data)
    csv_file = f"./data/match_squad_{datetime.now().strftime('%Y-%m-%d')}.csv"
    df.to_csv(csv_file, index=False)
    return df

match_squad_df = match_squad_info(api_key_4,matches_df)
```
